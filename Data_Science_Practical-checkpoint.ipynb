2. IQR (Interquartile Range) Method:

The IQR method defines an outlier as a data point that falls below the first quartile (Q1) minus 1.5 times the interquartile range (IQR), or above the third quartile (Q3) plus 1.5 times the IQR. The IQR is the range between the first and third quartiles.

IQR = Q3 ‚àí Q1

LowerBoundary = Q1 ‚àí 1.5 √ó IQR

3. Box Plots:

Description: Box plots provide a visual representation of the distribution of data. They show the median, quartiles, and potential outliers. Points beyond the whiskers (calculated based on IQR) are considered outliers. 
For example; Suppose we have a box plot with whiskers at 1.5 times the IQR. If data points extend beyond these whiskers, they are identified as outliers.
